## Load dataset
read the data from csv file,set an index and convert some columns to category type

```python
data=pd.read_csv('data.csv')
data.set_index('shot_id',inplace=True)  #set an index ,named as 'shot_id'
# collect the columns which should be converted to categoy type
colnames_list1=['combined_shot_type','game_event_id',
               'game_id','playoffs',
                'season','shot_made_flag',
              'shot_type','team_id']  
#collect the columns which should be converted to object type
colnames_list2=['action_type','period']
for colname in colnames_list1:
    data[colname]=data[colname].astype('category')
for colname in colnames_list2:
    data[colname]=data[colname].astype('object')
```

## Data Cleaning
Assuming an independence of each shot,therefore some columns might be dropped 

```python
data_c1=data.copy()  # create a copy of data frame
target=data_c1['shot_made_flag'].copy()
#Remove some columns
data_c1.drop('team_id',axis=1,inplace=True)   #Always one number
data_c1.drop('lat',axis=1,inplace=True)       #Correlated with loc_x
data_c1.drop('lon',axis=1,inplace=True)       #Correlated with loc_y
data_c1.drop('game_id',axis=1,inplace=True)   #Independent
data_c1.drop('game_event_id',axis=1,inplace=True)  #Always LA Lakers
data_c1.drop('shot_made_flag',axis=1,inplace=True)
```

There is still some outliers which should be removed.

```python
def detect_outliers(series,whis=1.5):
    q75,q25=np.percentile(series,[75,25])
    iqr=q75-q25
    return(~((series-series.median()).abs()<=(whis*iqr)))
```

## Data Transformation
### New features

```python
data_c1['action_type'].value_counts().sort_values()
```

```python
data_c1['action_type'].value_counts().sort_values().index
```

```python
#Remaining time
data_c1['seconds_from_period_end']=60*data_c1['minutes_remaining']+data_c1['seconds_remaining']
data_c1['last_5_sec_in_period']=data_c1['seconds_from_period_end']<5
data_c1.drop('minutes_remaining',axis=1,inplace=True)
data_c1.drop('seconds_remaining',axis=1,inplace=True)
data_c1.drop('seconds_from_period_end',axis=1,inplace=True)

#Matchup -(away/home)
data_c1['home_play']=data_c1['matchup'].str.contains('vs').astype('int')
data_c1.drop('matchup',axis=1,inplace=True)

#Game date
data_c1['game_date']=pd.to_datetime(data_c1['game_date'])
data_c1['game_year']=data_c1['game_date'].dt.year
data_c1['game_month']=data_c1['game_date'].dt.month
data_c1.drop('game_date',axis=1,inplace=True)

#Loc_x,and loc_y binning
data_c1['loc_x']=pd.cut(data_c1['loc_x'],25)
data_c1['loc_y']=pd.cut(data_c1['loc_y'],25)

#Replace 20 least common action types with value 'Other'
rare_action_types=data_c1['action_type'].value_counts().sort_values().index.values[:20]
data_c1.loc[data_c1['action_type'].isin(rare_action_types),'action_type']='Other'
```

#### Encode categorical variables

```python
pd.get_dummies(data_c1['action_type']).add_prefix('{}#'.format('action_type'))
```

```python
categorial_cols = [
    'action_type', 'combined_shot_type', 'period', 'season', 'shot_type',
    'shot_zone_area', 'shot_zone_basic', 'shot_zone_range', 'game_year',
    'game_month', 'opponent', 'loc_x', 'loc_y']
for cc in categorial_cols:
    dummies=pd.get_dummies(data_c1[cc])
    dummies=dummies.add_prefix('{}#'.format(cc))
    data_c1.drop(cc,axis=1,inplace=True)
    data_c1=data_c1.join(dummies)
```
